```python
import pandas as pd
import matplotlib.pyplot as plt         
import numpy as np
import random

df=pd.read_csv("../data/world_bank_development_indicators.csv")
df_2020=df[df["date"].astype(str).str.contains("2020")]

data=df_2020["life_expectancy_at_birth"].dropna().tolist()#dropna() removes missing values, tolist() converts to list

data=sorted(data)
# now we compute the edf for the full data set (treated as true edf)
true_y=[(i+1)/len(data) for i in range(len(data)) ]

sammple_size=[10,30,100]
plt.figure(figsize=(8,6))  #IT JUST SETS THE FIGURE SIZE

for n in sammple_size:
    sample=sorted(random.sample(data,n))#random.sample samples n values from data(for example chooses random 10 countries from the data and sorts them)
    edf_y=[(i+1)/len(sample) for i in range(len(sample)) ]#computes the edf for the sample
    plt.step(sample,edf_y, where='post', label="edf for sample")

    plt.step(data,true_y, where='post', label="true edf", color='black', linewidth=2)

    plt.xlabel("Life Expectancy at Birth (years)")
    plt.ylabel("Cumulative Proportion of Countries")
    plt.title("Glivenko-Cantelli Theorem Demonstration:EDF convergence with Increasing Sample Size")
    plt.legend()
    plt.grid(True)
    plt.show()
```

This theorem guarantees that as our dataset grows, the empirical distribution grows, the empirical distribution we observe(EDF) gets closer to the true population distribution.
In global development analytics, this means as more countries report data, our insights(like the distribution of life expectancy become more accurate and stable).
