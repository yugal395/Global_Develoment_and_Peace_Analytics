```python
import pandas as pd
import matplotlib.pyplot as plt
import random

df=pd.read_csv("../data/world_bank_development_indicators.csv")
df_2020=df[df['date'].astype(str).str.contains("2020")]

data=df_2020["life_expectancy_at_birth"].dropna().tolist() #dropna.values removes NaN values and converts to numpy array
true_mean=sum(data)/len(data)
print("True Mean:", true_mean)
sample_sizes=[5,10,30,50,100,120,150,170,190,200]
sample_means=[]
for n in sample_sizes:
    sample=random.sample(data,n) #takes randomly n countries from sample_sizes
    sample_mean=sum(sample)/len(sample)
    sample_means.append(sample_mean)
    print(f"Sample size {n} → average life expectancy: {sample_mean}")
```

```python
plt.plot(sample_sizes, sample_means, marker='o', label='Sample Means')
plt.axhline(y=true_mean, color='r', linestyle='--', label='True Mean')
plt.xlabel("No of countries in sample")
plt.ylabel("Average life expectancy at birth")
plt.title("Sample Means vs True Mean of Life Expectancy at Birth (2020)")
```

So with the concept of stochastic convergence if we take the sample size to be large enough, it should eventually converge to the true mean (true value) which is known as stochastic convergence and can be shown by this line graph that shows with increase in countries in sample the, average mean starts to converge to the true mean denoted by red line
