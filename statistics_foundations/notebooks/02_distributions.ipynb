```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/world_bank_development_indicators.csv")
df.head()
```

```python
sns.histplot(df["life_expectancy_at_birth"],kde=True)
plt.title("Distribution of life expectancy at birth")
plt.xlabel("Life Expectancy at Birth (years)")
plt.ylabel("No of countries")
plt.show()
```

```python
sns.histplot(df["political_stability_estimate"], kde=True)
plt.title("Distribution of political stability estimate")
plt.xlabel("Governance Score")
plt.ylabel("No of countries")
plt.show()
```

```python
sns.jointplot(data=df, x="GDP_current_US", y="life_expectancy_at_birth", kind="scatter")
plt.show()
```

```python
cols=["GDP_current_US", "life_expectancy_at_birth", "government_expenditure_on_education%", "political_stability_estimate"]
sns.heatmap(df[cols].corr(),annot=True, cmap="coolwarm")# annot =True to display correlation values, and cmap to set color scheme
plt.title("Correlation Heatmap among different indicators")
plt.show()
```

This graph represents how key indicators are distributed
* Histogram shows the spead and skewness of variables like life expectancy and political stability
* Jointplot visualises the link between wealth and well being
* Correlation matrix summaries the broader relationships and compares 4 indicators within a single snapshot
