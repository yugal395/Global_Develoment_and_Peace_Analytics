```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

df=pd.read_csv("../data/world_bank_development_indicators.csv")
df_2020=df[df["date"].astype(str).str.contains("2020")]

#picking two columns for modelling
x=df_2020["GDP_current_US"]
y=df_2020["life_expectancy_at_birth"]

data=df_2020[["GDP_current_US","life_expectancy_at_birth"]].dropna()
data["logGDP"]=np.log(data["GDP_current_US"])#log transformation to handle skewness in GDP data
slope,intercept,r_value,p_value,std_err=stats.linregress(data["GDP_current_US"],data["life_expectancy_at_birth"]) #slope gives the rate of change of life expectancy with GDP, intercept gives the life expectancy when GDP is 0, rvalue indicates strength and direction of relationship, p_value tests significance, std_err gives standard error of the estimate

data["predicted"]=intercept + slope*data["log_GDP"]

plt.scatter(data["GDP_current_US"],data["life_expectancy_at_birth"],label="Actual Data",alpha=0.5)# actual data points
plt.plot(data["GDP_current_US"],data["predicted_life_expectancy"],color="red",label="Model(fit)")# regression line
plt.xlabel("GDP (current US$)")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Statistical Modeling: GDP vs Life Expectancy (2020)")
plt.legend()
plt.grid(True)
plt.show()  

print("Slope(b):", slope)
print("Intercept(a):", intercept)
print("R-squared:", r_value**2)
```
