```python
import pandas as pd
import numpy as np
df=pd.read_csv("..\data\world_bank_development_indicators.csv")
df["date"]=pd.to_datetime(df["date"])  # we convert it to standard datetime format
df["year"]=df["date"].dt.year # extract year from date
df_2020=df[df["year"]==2020]# we filter data for year 2020
```

```python
life=df_2020['life_expectancy_at_birth'].dropna()
life.head()
```

```python
true_mean=life.mean()
true_mean


```

```python
sample_means=[]
for i in range(100):
    small_sample=life.sample(n=20, replace=True) 
    sample_means.append(small_sample.mean())
np.mean(sample_means)
```

If the sample size is 20, the estimate jumps around but still pretty close to the true mean indicating unbiasedness
but the average of many sample mean almost = full population mean

```python
means_10=[life.sample(n=10, replace=True).mean() for _ in range(1000)]
means_50=[life.sample(n=50, replace=True).mean() for _ in range(1000)]
means_100=[life.sample(n=100, replace=True).mean() for _ in range(1000)]



np.mean(means_10), np.mean(means_50), np.mean(means_100)
```

```python
np.std(means_10), np.std(means_50), np.std(means_100)
```

So, for topic 3.1
The sample mean is an unbiased estimator of true population mean.
We used our 2020 data for 'life expectancy at birth'
Simulation showed:
the average of many samples is very close to the true mean (unbiasedness)
as we increase sample size (10->50->100), the variation decreases(consistency)
