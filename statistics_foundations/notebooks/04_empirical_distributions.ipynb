```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("../data/world_bank_development_indicators.csv")
df_2020=df[df["date"].astype(str).str.contains("2020")]

data=df_2020["life_expectancy_at_birth"].dropna().tolist()#dropna() removes missing values, tolist() converts to list
```

```python
data_sorted=sorted(data)
#we sort the data to make it easier to compute the empirical CDF

y=[]
for i in range(len(data_sorted)):
    proportion=(i+1)/len(data_sorted)# Basically here we just created proportions from 1 country ton countries
    y.append(proportion)
```

```python
plt.step(data_sorted, y, where="post",label="2020",color="blue")
plt.xlabel("Life Expectancy at Birth (years)")
plt.ylabel("Cumulative Proportion of Countries")
plt.title("Empirical Distribution Function of Life Expectancy (2020)")
plt.grid(True) #Add grid for better readability
plt.step(data_2000,y_2000,where="post",label="2000",color="orange")
plt.legend()#Add legend to differentiate between years
plt.show()
```
